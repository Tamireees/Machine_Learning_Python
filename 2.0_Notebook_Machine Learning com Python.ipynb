### Parte 9: Gráfico de Densidade   /// Teste de normalidade 

Um gráfico de densidade, ou gráfico de estimativa de densidade de Kernel (KDE, do inglês Kernel Density Estimation), é uma representação gráfica que mostra a distribuição de dados em uma forma suave. Em vez de representar a distribuição de dados em barras discretas, como em um histograma, um gráfico de densidade usa uma linha contínua para estimar a densidade de probabilidade da variável.

Aqui está uma breve explicação dos principais pontos:

* Distribuição Suave: Ao contrário de um histograma, que mostra a frequência de valores em intervalos discretos, um gráfico de densidade cria uma curva suave que representa a distribuição dos dados ao longo de um intervalo contínuo. Isso ajuda a visualizar a tendência geral e a forma da distribuição.

* Estimação de Densidade de Kernel (KDE): A KDE é uma técnica que utiliza funções kernel para estimar a densidade de probabilidade de uma variável aleatória. Isso significa que ela calcula a densidade em torno de cada ponto de dados e soma essas densidades para criar a curva final.

* Parâmetro de Suavização (Bandwidth): A largura de banda (bandwidth) é um parâmetro crucial na KDE. Ela controla o grau de suavização da curva. Um bandwidth pequeno resulta em uma curva mais detalhada (menos suave), enquanto um bandwidth maior resulta em uma curva mais suave. Escolher o bandwidth adequado é importante para obter uma boa representação da distribuição dos dados.

* Visualização Intuitiva: Gráficos de densidade são intuitivos e facilitam a comparação de distribuições entre diferentes conjuntos de dados. Eles são úteis para identificar modos (picos), assimetrias e caudas na distribuição.

Interpretação do Gráfico de Densidade

* Picos (Modos): Os picos na curva representam os valores mais frequentes na distribuição.
* Largura da Curva: A largura da curva indica a variabilidade dos dados. Uma curva mais larga indica maior variabilidade.
* Assimetria: Se a curva não é simétrica, isso indica que a distribuição dos dados é assimétrica.
* Caudas: As extremidades da curva (caudas) mostram a presença de valores extremos (outliers).


Os gráficos de densidade são uma ferramenta poderosa para entender a distribuição e as características dos dados de forma mais detalhada e contínua do que um histograma tradicional.

```python
plt.figure(figsize=(10,5))

sns.kdeplot(data=df_studio, label="Albuns de Studio", fill=True)
sns.kdeplot(data=df_ao_vivo, label="Albuns Ao Vivo", fill=True)

plt.title("Distribuição do barulhos dos Albuns")
plt.xlabel("Barulho")
plt.ylabel("Densidade")

plt.legend()
plt.show()
```

* Teste de normalidade / Hipótese de Normalidade

Analisar o teste de normalidade e o valor-p é essencial para a análise de dados porque:

* Teste de Normalidade: Verifica se os dados seguem uma distribuição normal. Isso é importante porque muitos testes estatísticos assumem normalidade dos dados para serem válidos.

* Valor-p: Indica a probabilidade de obter os resultados observados se a hipótese nula for verdadeira. Um valor-p baixo (geralmente < 0.05) sugere que os resultados são estatisticamente significativos, rejeitando a hipótese nula.

Em resumo, esses testes ajudam a validar suposições estatísticas e a determinar a significância dos resultados, garantindo a precisão e a validade das conclusões tiradas dos dados.

```python
from scipy.stats import shapiro
```

```python
stat, p = shapiro(df_studio)

print("Soma do barulho dos albuns de Studio: ")
print('Estatística de teste: {:.4f}, valor p: {}'.format(stat, p))

if p > 0.05:
    print("Não há evidência suficiente para rejeitar a hipótese de Normalidade.")
else:
    print("A hipótese de normalidade é rejeitada")
```

```python
stat, p = shapiro(df_ao_vivo)

print("Soma do barulho dos albuns Ao Vivo: ")
print('Estatística de teste: {:.4f}, valor p: {}'.format(stat, p))

if p > 0.05:
    print("Não há evidência suficiente para rejeitar a hipótese de Normalidade.")
else:
    print("A hipótese de normalidade é rejeitada")
```

Normalização e padronização são técnicas importantes em análise de dados e aprendizado de máquina para preparar os dados antes de aplicar modelos. Mesmo que as variáveis não tenham uma relação direta, essas técnicas são frequentemente necessárias para melhorar a performance e a interpretação dos modelos. Aqui está o porquê:

* Normalização
Normalização, ou escalonamento, é o processo de ajustar os valores numéricos para que eles fiquem em uma faixa específica, geralmente entre 0 e 1. Isso é feito subtraindo o valor mínimo e dividindo pela diferença entre o valor máximo e o mínimo (ou usando outras técnicas de escalonamento).

 Razões para Normalização:

1. Modelos Sensíveis à Escala: Alguns algoritmos de aprendizado de máquina, como K-means, K-nearest neighbors (KNN) e redes neurais, são sensíveis à escala das variáveis. Variáveis com maiores magnitudes podem dominar a função objetivo e influenciar desproporcionalmente os resultados.
2. Melhoria da Convergência: Algoritmos de otimização, como gradiente descendente usado em redes neurais, convergem mais rapidamente quando as características são escalonadas de forma similar.
3. Manter a Proporção Original: Ao normalizar, mantemos a proporção original dos dados, mas ajustamos a escala para que todas as variáveis contribuam de forma equilibrada.


* Padronização
Padronização é o processo de transformar os dados para que tenham uma média de 0 e um desvio padrão de 1. Isso é feito subtraindo a média e dividindo pelo desvio padrão de cada variável.

 Razões para Padronização:
1. Comparação de Variáveis: Padronizar os dados permite comparar variáveis que originalmente tinham diferentes unidades ou escalas.
2. Modelos Estatísticos: Muitos modelos estatísticos assumem que os dados seguem uma distribuição normal (ou aproximadamente normal). Padronizar pode ajudar a atender a essas suposições.
3. Algoritmos Baseados em Distância: Algoritmos que dependem de medidas de distância, como SVM e KNN, funcionam melhor quando as variáveis estão na mesma escala. Padronização ajuda a garantir que uma variável não domine outras devido à sua escala maior.
4. Mitigar o Efeito de Outliers: Padronizar pode ajudar a reduzir o impacto de outliers, pois os dados são escalonados com base no desvio padrão, suavizando a influência de valores extremos.


* Exemplos de Uso
1. K-means Clustering: Se os dados não forem normalizados ou padronizados, variáveis com maior magnitude podem influenciar desproporcionalmente a formação dos clusters.
2. Redes Neurais: Dados não escalonados podem fazer com que os gradientes durante a fase de treinamento se tornem muito grandes ou muito pequenos, dificultando a convergência do modelo.

* Resumo

Embora as variáveis possam não ter uma ligação direta, normalizar e padronizar os dados é fundamental para garantir que todos os algoritmos de aprendizado de máquina e técnicas de análise de dados funcionem de maneira eficiente e precisa. Essas técnicas ajudam a evitar problemas de escala, melhoram a convergência de algoritmos de otimização e asseguram que todas as variáveis contribuam de maneira equilibrada para os resultados do modelo.

O que o random_state faz?
Quando você usa a função train_test_split da Scikit-learn para dividir seus dados em conjunto de treinamento e teste, o computador usa números aleatórios para decidir quem vai para cada grupo.

Por que usar o random_state?
O random_state é a "semente" desse sorteio, que se você definir um valor para o random_state, tipo random_state=42, o sorteio sempre vai sair igual, toda vez que você rodar o código. Isso é super útil porque:

Reprodutibilidade: Se você ou alguém mais quiser repetir seu experimento, vai conseguir exatamente os mesmos resultados. É a forma de garantir que todo mundo use o mesmo sorteio.
Consistência: Com o random_state fixo, você sabe que os resultados dos seus testes (tipo acurácia, precisão, etc.) não vão mudar por causa da divisão dos dados.

E a diferença entre valores diferentes?
Na verdade, o número que você escolhe para o random_state não importa muito. Pode ser 1, 42, 123, qualquer coisa. A única coisa que muda é o "padrão" dos números aleatórios gerados.

### Parte 4 / 5: Modelos de Regressão

Um modelo de Regressão linear é a técnica estatística usada para modelar o relacionamento entre uma variável dependente contínua e uma ou mais variáveis independentes. A regressão linear assume que existe uma relação linear entre as variáveis envolvidas, ou seja, a relação pode ser aproximada por uma linha reta.

```python
# Criando um objeto de Regressão Linear:
lr = LinearRegression()
```

```python
# X contem as variáveis preditoras ou independentes
X = imoveis[['Area', 'Suites', 'IA', 'Semruido', 'Vista', 'Andar', 'AV100m', 'DistBM']]

# y variável target ou dependente
y = imoveis[['Valor']]
```

Podemos ter alguns impedimentos que podem ocasionar resultados ruins em nossos algoritmos, sendo eles: Algoritmos ruins ou dados ruins.

Vamos conversar um pouquinho sobre esses desafios! Se tratando de dados ruins, podemos ter alguns fatores:

Quantidade insuficiente de dados: Com uma amostra de dados muito pequena, existirá um "ruído de amostragem" e se houver uma amostra muito grande com dados não representativos, o método de amostragem também pode ser falho (Viés de amostragem).

Dados de treinamento não representativos: O objetivo de um bom modelo de aprendizado de máquina é generalizar bem a partir dos dados de treinamento, sendo assim, é importante buscar uma base de dados representativa. Será que a sua base de dados consegue generalizar?

Dados de baixa qualidade: Aqui é preciso se dedicar a limpeza dos dados, uma base não consistente pode impactar na detecção de padrões.

Características irrelevantes: Entra lixo, sai lixo. Atenção aos dados que entram no seu modelo! A dica aqui é a dedicação na etapa de feature engineering ou técnicas de redução de dimensionalidade.

* Equilibrando a base de dados

Para equilibrar a base de dados, podemos utilizar algumas técnicas como por exemplo a biblioteca resample do sklearn.utils, utilizando o oversampling.

O oversampling é o processo de duplicar ou criar novas amostras da classe minoritária, enquanto o undersampling é o processo de remover algumas amostras da classe majoritária.

Vamos analisar como nosso resultado pode ter um uprade com oversampling?

```python
from sklearn.utils import resample
```

```python
# Separando as classes majoritárias e minoritárias
df_majority = airlines[airlines.Class == 0]
df_minority = airlines[airlines.Class == 1]
```

```python

# Upsampling da classe minoritária
df_minority_upsampled = resample(df_minority, replace=True, n_samples=len(df_majority),  random_state=7)
```

```python
# Juntando os dois DataFrames
df_equilibrado = pd.concat([df_majority, df_minority_upsampled])
```

```python
sns.countplot(x='Class',data=df_equilibrado)
```

Agora sim, temos uma base de dados equilibrada!

Podemos tentar buscar novas alternativas utilizando outros algoritmos de machine learning.

Agora, vamos falar um pouquinho sobre algoritmos ruins:

* Sobreajuste nos dados (overfiting): Quando o seu modelo funciona muito bem com os dados de treinamento mas não generaliza bem novos dados de entrada. Isso pode acontecer quando o modelo é muito complexo em relação ao ruído e quantidade. Como solução podemos pensar aqui em:

1. Simplificar o modelo.
2. Coletar mais dados.
3. Reduzir o ruído (exemplo, remover outliers).
4. Regularização: Chamamos de regularização quando restringimos um modelo para simplificar e reduzir o risco de reajuste dos dados. A regularização pode ajudar a generalizar melhor o modelo em novos exemplos de dados.

* Subajuste dos dados (underfiting): Nesse caso seu modelo ficou muito simples ao ponto de não aprender corretamente os dados. Algumas soluções:

1. Selecionar um modelo mais poderoso.
2. Feature engineering.
3. Reduzir as regularizações.
